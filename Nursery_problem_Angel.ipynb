##Perfilamiento del DataSet
A continuación trataremos de explorar algo de información que pueda sernos de utilidad para abordar el problema con los modelos de ML adecuados.

```
CHARACTERISTIC_COLS = ['parents', 'has_nurs', 'form', 'children', 'housing', 'finance', 'social', 'health']
FULL_COLS = ['parents', 'has_nurs', 'form', 'children', 'housing', 'finance', 'social', 'health', 'evaluation']
ndf = pd.read_csv(NURSERY_INFO_DIR + 'nursery_with_headers.data') # Para que el dataset contenga las características y la variable objetivo
ndf.profile_report(style={'full_width':True})
```

```
dfs_with_dummie_values = {}
for column in ndf.columns:
  ndf[column] = pd.Categorical(ndf[column])
  dfs_with_dummie_values[column] = pd.get_dummies(ndf[column], prefix = column)
p_ndf = ndf
for df_key in dfs_with_dummie_values.keys():
  p_ndf = pd.concat([p_ndf, dfs_with_dummie_values[df_key]], axis=1)
p_ndf = p_ndf.drop(FULL_COLS, axis=1)
p_ndf
```
